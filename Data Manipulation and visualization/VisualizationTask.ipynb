```python
df= df.dropna(subset=['latitude_x','longitude_x','revenue_x','tier_y','hq_city_y','hq_country_y','hq_region_y','latitude_y',
                     'longitude_y','employees_y'])
```

```python
df.head()
```

```python
matrix = df.apply(lambda x: x.factorize()[0]).corr()
```

```python
matrix
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(matrix,yticklabels=True,cbar=True,cmap='coolwarm',annot=True)
```

```python
## in order to have better view i will drop columns and rows tier_y,hq_country_y,employees_y
```

```python
matrix = matrix.drop(['tier_y','hq_country_y'])
```

```python
matrix = matrix.drop(['tier_y','hq_country_y','employees_y'],axis=1)
```

```python
matrix = matrix.drop(['employees_y'])
```

```python
matrix
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(matrix,yticklabels=True,cbar=True,cmap='coolwarm')
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(matrix,yticklabels=True,cbar=True,cmap='coolwarm', annot=True)
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(matrix,yticklabels=True,cbar=True,cmap='plasma', annot=True)
```

```python
#as it is seen on the heatmaps the biggest correlation is between latittude_x, 
#longitude_x(which is location of a company) and 
#latittude_y, longitude_y(which is location of a vendor) . 
#So based on this it is possible to say that when company chooses a vendor
#loacation of this vendor plays a big role. I can assume that the closer the vendor is to 
#company the bigger chance is to be chosen.
```
