Ecommerce Customers csv file from the company. It has Customer info, suchas Email, Address, and their color Avatar. Then it also has numerical value columns:

Avg. Session Length: Average session of in-store style advice sessions.
Time on App: Average time spent on App in minutes
Time on Website: Average time spent on Website in minutes
Length of Membership: How many years the customer has been a member.

```python
X=df[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=101)
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
#9
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
coeff = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficent'])
coeff
```

```python
#10
predictions=lm.predict(X_test)
predictions
```

```python
scatter=sns.scatterplot(x=y_test,y=predictions)
scatter.set(ylim=(200,800),xlim=(200,800))
scatter.set_ylabel('Predicted Y')
scatter.set_xlabel('Y test')
```

```python
#9
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(y_test,predictions)
MSE = metrics.mean_squared_error(y_test,predictions)
RMSE = np.sqrt(MSE)
```

```python
print('MAE:',MAE )
print('MSE:',MSE )
print('RMSE:',RMSE )
```

```python
#10
Residuals = y_test - predictions
Residuals.head()
```

```python
hist=sns.distplot((y_test-predictions),kde=True,bins=50,color='brown')
hist.set(ylim=(0,0.06),xlim=(-40,50))
```
