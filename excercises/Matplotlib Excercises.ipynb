```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
#importing
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Create a figure object called fig using plt.figure()
#Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])


ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.ylim(0,200)
plt.xlim(0,100)

ax.plot(x,y)
```

```python
#Create a figure object and put two axes on it, ax1 and ax2. 
#Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively
```

```python
fig = plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
#Now plot (x,y) on both axes. And call your figure object to show it
```

```python
fig = plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
ax1.set_ylim(0,200)
ax1.set_xlim(0,100)
ax2.set_ylim(0,200)
ax2.set_xlim(0,100)

plt.show()
```

```python
#Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot
```

```python
fig = plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.4])

ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

ax1.plot(x,z)
ax1.set_title('zoom')
ax1.set_ylabel('z')
ax1.set_xlabel('x')
ax1.set_xlim(0,100)
ax1.set_ylim(0,10000)
```
