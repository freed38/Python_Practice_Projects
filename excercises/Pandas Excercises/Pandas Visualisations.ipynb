```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

```python
#SCATTER PLOT
```

```python
df3.plot.scatter(x='a',y='b',c='red',figsize=(12,3),s=50)
```

```python
#HISTOGRAM
plt.style.use('ggplot')
df3['a'].plot.hist()
```

```python
#STYLING
plt.style.use('ggplot')
df3['a'].plot.hist(bins=25,alpha=0.5,figsize=(6,4.5),lw=0.5,rwidth=0.89)
```

```python
#Create a boxplot comparing the a and b columns.
```

```python
df3[['a','b']].plot.box()
```

```python
#Create a kde plot of the 'd' column
```

```python
df3['d'].plot.density(lw=1,figsize=(7,7))
```

```python
#Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)
```

```python
df3['d'].plot.density(lw=5,linestyle='-.')
```

```python
#Create an area plot of all the columns for just the rows up to 30. (hint: use .ix)
```

```python
df3[0:31].plot.area(alpha=0.4,figsize=(10,5))
```

```python
#Remove legend out of the plot
```

```python
ax=df3[0:31].plot.area(alpha=0.4,lw=1,figsize=(10,5))
leg=ax.legend(bbox_to_anchor=(1.16,0.7))
```
