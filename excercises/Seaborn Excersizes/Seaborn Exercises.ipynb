```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
# JointPlot
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
# DISTPLOT
sns.distplot(titanic['fare'],kde=False,bins=30,color='red')
```

```python
#BOXPLOT
sns.boxplot(x='class',y='age',data=titanic,order=['First','Second','Third'],palette='rainbow')
```

```python
#SWARMPLOT
sns.swarmplot(x='class',y='age',data=titanic,order=['First','Second','Third'],palette='rainbow')
```

```python
#COUNTPLOT
sns.countplot(x='sex',data=titanic,palette='viridis')
```

```python
#HEATMAP
plt.figure(figsize=(8,8))
titanic.corr()
sns.heatmap(data=titanic.corr(),vmin=-0.8,vmax=0.8,cmap='coolwarm',annot=True).set(title='titanic.corr()')
```
